```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```




importing the required libraries

```
img = cv2.imread('/content/sample_data/images (1).jpg')
#cv2_imshow(img)
Z = img.reshape((-1,3))
```

The reshape() function is used to give a new shape to an array without changing its data. 
Array to be reshaped. The new shape should be compatible with the original shape
if we do not know the number of rowsto be given, for it to reshape so we give the parameter as -1,
when we give the parameter as -1 numpy will figure out by itself what should be the row size based on image dimensions and 3 is the number of columns


```
Z = np.float32(Z)
```

converts the image into a numpy array with values of type single precesion float.




```
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

criteria: It is the iteration termination criterion
this condition is met, the algorithm iteration stops
in fact, it should be three yuan set of parameters. They are（type，max_iter，epsilon）

cv2.TERM_CRITERIA_EPS - If the specified accuracies epsilon, stopping iterative algorithm.

cv.TERM_CRITERIA_MAX_ITER - stop the algorithm after a specified number of iterations max_iter.

cv.TERM_CRITERIA_EPS+ cv.TERM_CRITERIA_MAX_ITER - iteration stops when any of the above conditions are satisfied.

```
K = 3
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
#combines it into a single 1dimensional array
res2 = res.reshape((img.shape))
#reshaping the image to it's original shape
cv2_imshow(res2)
# used for displaying the image
cv2.imwrite('k=3.jpg',res2)
#used to write the same image onto the computer 
```

Z-data

K-number of clusters

bestLabels-None default classification labels 

max_iter - Specifies the maximum number of iterations integer
flags: This flag specifies the initial center by way of. cv2.KMEANS_RANDOM_CENTERS

cv2.KMEANS_RANDOM_CENTERS-With this flag enabled, the method always starts with a random set of initial samples, and tries to converge
